```python
import numpy as np
import pandas as pd

df=pd.read_csv("Iris.csv")
print(df)
```

```python
df1=df.drop(columns=["Id"],axis=1)  # axis=1 to remove the data in column wise.....#axis=0 to remove data in row wise
print(df1)
```

```python
df1.dtypes  # To know data types
```

```python
x=df1.drop("Species",axis=1).values  #To get input    #values enn kodthaaal...array ayt covert cheyyam
y=df1["Species"].values  #To get output
x
```

```python
y
```

```python
#Testing and training
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#Next process is normalisation
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_test
```

```python
x_train
```

```python
#Next process is model creation

from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=7)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
#Here we know y_pred=y_test so just print y_test
```

```python
y_test
```

```python
print(model.predict(scaler.transform([[6,3,5,7]])))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
#Practical method of accuracy_score
score=accuracy_score(y_test,y_pred)
score
```
